```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
ex1_B1 = {"Real": [60, 180, 120], "Measured": [74, 173, 145]}
ex1_B2 = {"Real": [120, 180, 60], "Measured": [116,195, 81]}
ex1_B1_df = pd.DataFrame(ex1_B1, index=["Waiting1", "Exam1", "Waiting2"])
ex1_B2_df = pd.DataFrame(ex1_B2, index=["Waiting1", "Exam1", "Waiting2"])
```

```python
n_groups = 3
index = np.arange(n_groups)
real_B1 = ex1_B1_df["Real"]
measured_B1 = ex1_B1_df["Measured"]
real_B2 = ex1_B2_df["Real"]
measured_B2 = ex1_B2_df["Measured"]
```

```python
fig, axes = plt.subplots(2, 1, sharex=True, sharey=True, figsize=(10, 7))
plt.rc('font', size=15) 
plt.rc('legend', fontsize=10)
bar_width = 0.35
opacity = 0.8
labels = ["Waiting 1", "Exam 1", "Waiting 2"]

axes[0].set_ylabel("Patient A")
axes[1].set_ylabel("Patient B")
axes[0].set_xticks(np.arange(len(labels))+bar_width/2)
axes[1].set_xticks(np.arange(len(labels))+bar_width/2)
axes[0].set_xticklabels(labels)
fig.text(0.5, 0.04, 'Experiment phases', ha='center')
fig.text(0.04, 0.5, 'Durations [seconds]', va='center', rotation='vertical')
plt.suptitle("Real vs Measured durations for Experiment 1")

rects1_1 = axes[0].bar(index, real_B1, bar_width,
alpha=opacity,
color='g',
label='Real Durations')
rects1_2 = axes[0].bar(index + bar_width, measured_B1, bar_width,
alpha=opacity,
color='r',
label='Measured Durations')
axes[0].legend()
axes[0].grid()

rects2_1 = axes[1].bar(index, real_B2, bar_width,
alpha=opacity,
color='g',
label='Real Durations')
rects2_2 = axes[1].bar(index + bar_width, measured_B2, bar_width,
alpha=opacity,
color='r',
label='Measured Durations')
axes[1].legend()
axes[1].grid()

plt.savefig("Ex1.png", dpi=300)
plt.show()
```
