```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
ex3_B1 = {"Real": [120, 120, 120, 120], "Measured": [132, 122, 100, 143]}
ex3_B2 = {"Real": [120, 120, 120, 120], "Measured": [128, 102, 117, 143]}
ex3_B3 = {"Real": [480, 0, 0, 0], "Measured": [515, 0, 0, 0]}
ex3_B1_df = pd.DataFrame(ex3_B1, index=["Waiting1", "Exam1", "Triage2", "Waiting2"])
ex3_B2_df = pd.DataFrame(ex3_B2, index=["Waiting1", "Triage1", "Exam2", "Waiting2"])
ex3_B3_df = pd.DataFrame(ex3_B3, index=["Waiting", "Waiting", "Waiting", "Waiting"])
```

```python
n_groups = 4
index = np.arange(n_groups)
real_B1 = ex3_B1_df["Real"]
measured_B1 = ex3_B1_df["Measured"]
real_B2 = ex3_B2_df["Real"]
measured_B2 = ex3_B2_df["Measured"]
real_B3 = ex3_B3_df["Real"]
measured_B3 = ex3_B3_df["Measured"]
```

```python
fig, axes = plt.subplots(3, 1, figsize=(15, 15))
plt.rc('font', size=15) 
plt.rc('legend', fontsize=10)
bar_width = 0.2
opacity = 0.8
labels1 = ["Waiting 1", "Exam 1", "Triage2", "Waiting 2"]
labels2 = ["Waiting 1", "Triage1", "Exam 2",  "Waiting 2"]

axes[0].set_ylabel("Patient A")
axes[1].set_ylabel("Patient B")
axes[2].set_ylabel("Patient C")
axes[0].set_xticks(np.arange(len(labels1))+bar_width/2)
axes[1].set_xticks(np.arange(len(labels1))+bar_width/2)
axes[2].set_xticks(np.arange(1)+bar_width/2)
axes[0].set_xticklabels(labels1)
axes[1].set_xticklabels(labels2)
axes[2].set_xticklabels(["Waiting"])
fig.text(0.5, 0.04, 'Experiment phases', ha='center')
fig.text(0.04, 0.5, 'Durations [seconds]', va='center', rotation='vertical')
plt.suptitle("Real vs Measured durations for Experiment 3")

rects1_1 = axes[0].bar(index, real_B1, bar_width,
alpha=opacity,
color='g',
label='Real Durations')
rects1_2 = axes[0].bar(index + bar_width, measured_B1, bar_width,
alpha=opacity,
color='r',
label='Measured Durations')
axes[0].legend()
axes[0].grid()

rects2_1 = axes[1].bar(index, real_B2, bar_width,
alpha=opacity,
color='g',
label='Real Durations')
rects2_2 = axes[1].bar(index + bar_width, measured_B2, bar_width,
alpha=opacity,
color='r',
label='Measured Durations')
axes[1].legend()
axes[1].grid()

rects3_1 = axes[2].bar(index, real_B3, bar_width,
alpha=opacity,
color='g',
label='Real Durations')
rects3_2 = axes[2].bar(index + bar_width, measured_B3, bar_width,
alpha=opacity,
color='r',
label='Measured Durations')
axes[2].legend()
axes[2].grid()

plt.savefig("Ex3.png", dpi=300)
plt.show()
```
