```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
ex_1 = {"Real": [60, 120, 180, 180, 120, 60], "Measured": [74, 116, 173, 195, 145, 81]}
ex_1_df = pd.DataFrame(ex_1, index=["Waiting1_B1", "Waiting1_B2", "Exam1_B1", "Exam1_B2", "Waiting2_B1", "Waiting2_B2"])
ex_1_df.head()
```

```python
n_groups = 6
real = ex_1_df["Real"]
measured = ex_1_df["Measured"]
errors = (real - measured)/2
errors
```

```python
# create plot
fig, ax = plt.subplots(figsize=(10, 5))
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, real, bar_width,
alpha=opacity,
color='g',
label='Real Durations')

rects2 = plt.bar(index + bar_width, measured, bar_width,
alpha=opacity,
color='r',
label='Measured Durations')

plt.xlabel('Experiment Phases')
plt.ylabel('Durations [seconds]')
plt.title('Real vs Measured Durations from Experiment 1')
plt.xticks(index + bar_width/2, (ex_1_df.index))
plt.legend()

plt.tight_layout()
plt.show()
```

```python
n_groups = 3
B1 = ex_1_df.loc[["Waiting1_B1", "Exam1_B1", "Waiting2_B1"]]["Real"]
B2 = ex_1_df.loc[["Waiting1_B2", "Exam1_B2", "Waiting2_B2"]]["Real"]
errors = (real - measured)/2
errors_B1 = [errors[0], errors[2], errors[4]]
errors_B2 = [errors[1], errors[3], errors[5]]
```

```python
# create plot
fig, ax = plt.subplots(figsize=(10, 5))
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8

rects1 = plt.bar(index, B1, yerr=errors_B1, width=bar_width,
alpha=opacity,
color='g',
label='B1',
ecolor="black",
capsize=5)

rects2 = plt.bar(index + bar_width, B2, yerr=errors_B2, width=bar_width,
alpha=opacity,
color='r',
label='B2',
ecolor="black",
capsize=5)

plt.xlabel('Experiment Phases')
plt.ylabel('Durations [seconds]')
plt.title('Real vs Measured Durations from Experiment 1')
plt.xticks(index + bar_width/2, (["Waiting1", "Exam1", "Waiting2"]))
plt.legend()

ax.grid()
plt.tight_layout()
plt.show()
```
